# Clean the dataset

```python

# drop the unit number column, more than five students column 
Student_addresses_2020_2021 = Student_addresses_2020_2021.drop(columns=['6d. unit #', '9. 5 or more undergrads/unit (y/n)'])
Student_addresses_2021_2022 = Student_addresses_2021_2022.drop(columns=['6d. unit #', '9. 5 or more undergrads/unit (y/n)'])
Student_addresses_2022_2023 = Student_addresses_2022_2023.drop(columns=['6d. unit #', '9. 5 or more undergrads/unit (y/n)'])
Student_addresses_2023_2024 = Student_addresses_2023_2024.drop(columns=['6d. unit #', '9. 5 or more undergrads/unit (y/n)'])

#Add zeros to the zipcodes in each student dataset
Student_addresses_2020_2021["6e. zip"] = "0" + Student_addresses_2020_2021["6e. zip"]
Student_addresses_2021_2022["6e. zip"] = "0" + Student_addresses_2021_2022["6e. zip"]
Student_addresses_2022_2023["6e. zip"] = "0" + Student_addresses_2022_2023["6e. zip"]
Student_addresses_2023_2024["6e. zip"] = "0" + Student_addresses_2023_2024["6e. zip"]

#Create a street combined column for the API code for each dataset
Student_addresses_2020_2021["Street_combined"] = Student_addresses_2020_2021['6a. street #']+ " " + Student_addresses_2020_2021['6b. street name'] + " " +Student_addresses_2020_2021['6c. street suffix']
Student_addresses_2021_2022["Street_combined"] = Student_addresses_2021_2022['6a. street #']+ " " + Student_addresses_2021_2022['6b. street name'] + " " +Student_addresses_2021_2022['6c. street suffix']
Student_addresses_2022_2023["Street_combined"] = Student_addresses_2022_2023['6a. street #']+ " " + Student_addresses_2022_2023['6b. street name'] + " " +Student_addresses_2022_2023['6c. street suffix']
Student_addresses_2023_2024["Street_combined"] = Student_addresses_2023_2024['6a. street #']+ " " + Student_addresses_2023_2024['6b. street name'] + " " +Student_addresses_2023_2024['6c. street suffix']

#Create a city and state column for each dataset
Student_addresses_2020_2021['City'] = "Boston"
Student_addresses_2020_2021['State'] = "Massachusetts"

Student_addresses_2021_2022['City'] = "Boston"
Student_addresses_2021_2022['State'] = "Massachusetts"

Student_addresses_2022_2023['City'] = "Boston"
Student_addresses_2022_2023['State'] = "Massachusetts"

Student_addresses_2023_2024['City'] = "Boston"
Student_addresses_2023_2024['State'] = "Massachusetts"

# This function will refill the street suffix column with the correct suffixs and remove duplicates
def correct_street_suffix_and_duplicates(df):
    # ---- ensure columns exist ----
    for col in ['6b. street name', '6c. street suffix']:
        if col not in df.columns:
            df[col] = pd.NA

    # ---- list of suffixes (common full + abbreviations + punctuation) ----
    valid_suffixes = [
        'Street', 'St', 'St.', 'Avenue', 'Ave', 'Ave.', 'Boulevard', 'Blvd', 'Blvd.',
        'Road', 'Rd', 'Rd.', 'Drive', 'Dr', 'Dr.', 'Court', 'Ct', 'Ct.',
        'Lane', 'Ln', 'Ln.', 'Place', 'Pl', 'Pl.', 'Terrace', 'Ter', 'Ter.',
        'Way', 'Wy', 'Wy.', 'Circle', 'Cir', 'Cir.', 'Parkway', 'Pkwy', 'Pkwy.',
        'Square', 'Sq', 'Sq.', 'Trail', 'Trl', 'Trl.', 'Loop', 'Lp', 'Lp.',
        'Highway', 'Hwy', 'Hwy.', 'Crescent', 'Cres', 'Cres.'
    ]

    # ---- regex pattern to find suffix at end ----
    pattern = r'\b(' + '|'.join(valid_suffixes) + r')\.?$'

    # ---- prepare clean street names ----
    s = df['6b. street name'].fillna('').astype(str).str.strip()

    # ---- extract suffixes ----
    extracted = s.str.extract(pattern, flags=re.IGNORECASE, expand=False)
    extracted = extracted.str.replace('.', '', regex=False).str.title().replace('', pd.NA)

    # ---- fill NaN suffixes with extracted ones ----
    df['6c. street suffix'] = (
        df['6c. street suffix']
        .fillna(extracted)
        .astype('string')
        .str.replace('.', '', regex=False)
        .str.title()
    )

    # ---- remove suffix from street name ----
    cleaned = s.str.replace(pattern, '', flags=re.IGNORECASE, regex=True)
    cleaned = cleaned.str.replace(r'\s+', ' ', regex=True).str.strip()
    cleaned = cleaned.replace('', pd.NA).astype('string').str.title()
    df['6b. street name'] = cleaned

    # ---- normalize suffix capitalization ----
    df['6c. street suffix'] = df['6c. street suffix'].str.replace('.', '', regex=False).str.title()

    # ---- drop duplicate addresses ----
    dup_cols = ['6a. street #', '6b. street name', '6c. street suffix', '6e. zip']
    df = df.drop_duplicates(subset=dup_cols, keep='first').reset_index(drop=True)

    return df


```

```python
# Call above function on all the address datasets
Student_addresses_2020_2021 = correct_street_suffix_and_duplicates(Student_addresses_2020_2021)
Student_addresses_2021_2022 = correct_street_suffix_and_duplicates(Student_addresses_2021_2022)
Student_addresses_2022_2023 = correct_street_suffix_and_duplicates(Student_addresses_2022_2023)
Student_addresses_2023_2024 = correct_street_suffix_and_duplicates(Student_addresses_2023_2024)


# How many rows they have now after dropping duplicate rows
print(Student_addresses_2020_2021.shape)
print(Student_addresses_2021_2022.shape)
print(Student_addresses_2022_2023.shape)
print(Student_addresses_2023_2024.shape)
```

```python

# This function will focus on deleting rows that have NA in street or suffix, and account for mispellings
def extensive_suffix_column_cleaning(df):
    # dictionary: keys are messy variants, values are correct suffixes
    suffix_map = {
        # common street suffixes
        'St': 'Street', 'St.': 'Street', 'St,': 'Street', 'St.,': 'Street',
        'Steet': 'Street', 'Stree': 'Street', 'Stret': 'Street', 'Sreet': 'Street',
        
        'Ave': 'Avenue', 'Ave.': 'Avenue', 'Ave,': 'Avenue', 'Avenu': 'Avenue',
        'Avenue,': 'Avenue', 'Aveue': 'Avenue', 'Aveneue': 'Avenue', 'Av': 'Avenue', 'Avene': 'Avenue',
        
        'Rd': 'Road', 'Rd.': 'Road', 'Road,': 'Road', 'Roa': 'Road', 'Driv': 'Drive',
        
        'Dr': 'Drive', 'Dr.': 'Drive', 'Dr,': 'Drive', 'Drive,': 'Drive',
        
        'Blvd': 'Boulevard', 'Blvd,': 'Boulevard', 'Bvd': 'Boulevard',
        
        'Ct': 'Court', 'Ct.': 'Court',
        
        'Lane': 'Lane', 'Ln': 'Lane',
        
        'Terr': 'Terrace', 'Terrace': 'Terrace', 'Terr.': 'Terrace', 'Terrance': 'Terrace',
        
        'Circle': 'Circle', 'Cir': 'Circle', 'Cct': 'Circle',
        
        'Place': 'Place', 'Place.': 'Place', 'Place,': 'Place', 'Pl': 'Place', 'Pl.': 'Place', 'Pl.,': 'Place', 'Plz': 'Place',
        
        'Way': 'Way', 'Way.': 'Way',
        
        'Parkway': 'Parkway', 'Pkwy': 'Parkway', 'Pky': 'Parkway',
        
        'Square': 'Square', 'Sq': 'Square',
        
        # optional: directions if you want to standardize
        'E': 'East', 'W': 'West', 'S': 'South', 'N': 'North', 'East': 'East', 'West': 'West',
    }


    ### 2. Map your `6c. street suffix` column using this dictionary

    # normalize: remove extra whitespace and title case
    df['6c. street suffix'] = df['6c. street suffix'].astype(str).str.strip().str.replace(r'[\s\.]+$', '', regex=True).str.title()

    # map variants to canonical suffix
    df['6c. street suffix'] = df['6c. street suffix'].map(suffix_map).fillna(pd.NA)


    ### 3. remove clearly invalid entries

    # remove numbers, apartment/floor info, city names, etc.
    invalid_suffixes = ['#7', 'Floor 3', 'Apt', 'Apartment', 'Boston', 'Brookline', 'Berkeley', 'Charlestown', 'Longwood', 'Stephen', 'Edward']

    df.loc[df['6c. street suffix'].isin(invalid_suffixes), '6c. street suffix'] = pd.NA
    df.loc[df['6b. street name'].isin(invalid_suffixes), '6c. street suffix'] = pd.NA

    # Drop rows where either street name or suffix is NA
    df = df.dropna(subset=['6b. street name', '6c. street suffix'])

    return df
```

# API code for filling in the lat and long columns of each dataset

```python
import time
import requests
import pandas as pd
import numpy as np
from typing import Optional, Tuple, Dict

GEOCODE_URL = "https://geocode.maps.co/search"

def _build_address_string(street: Optional[str], city: Optional[str],
                          state: Optional[str], postalcode: Optional[str]) -> str:
    parts = []
    if street and str(street).strip():
        parts.append(str(street).strip())
    if city and str(city).strip():
        parts.append(str(city).strip())
    if state and str(state).strip():
        parts.append(str(state).strip())
    if postalcode and str(postalcode).strip():
        parts.append(str(postalcode).strip())
    return ", ".join(parts)

def geocode_single(session: requests.Session,
                    street: Optional[str],
                    city: Optional[str],
                    state: Optional[str],
                    postalcode: Optional[str],
                    api_key: Optional[str] = None,
                    country: Optional[str] = "US",
                    timeout: float = 10.0,
                    max_retries: int = 4,
                    pause: float = 1.0) -> Tuple[Optional[float], Optional[float]]:
    """
    Geocode a single address. Returns (lat, lon) or (None, None) if no result.
    Uses exponential backoff for 429/503 responses.
    """
    address_str = _build_address_string(street, city, state, postalcode)
    if not address_str:
        return None, None

    params = {
        "format": "json",
        "limit": 1
    }
    # Prefer structured search: include components when available
    if street:
        params["street"] = street
    if city:
        params["city"] = city
    if state:
        params["state"] = state
    if postalcode:
        params["postalcode"] = postalcode
    if country:
        params["country"] = country

    # If user prefers to pass api_key as query param, uncomment next line
    # if api_key: params["api_key"] = api_key

    headers = {
        "User-Agent": "geocode-client/1.0"
    }
    if api_key:
        headers["Authorization"] = f"Bearer {api_key}"

    backoff = pause
    for attempt in range(1, max_retries + 1):
        try:
            resp = session.get(GEOCODE_URL, params=params, headers=headers, timeout=timeout)
        except requests.RequestException:
            # network failure
            if attempt == max_retries:
                return None, None
            time.sleep(backoff)
            backoff *= 2
            continue

        if resp.status_code == 200:
            try:
                data = resp.json()
            except ValueError:
                return None, None
            if isinstance(data, list) and len(data) > 0:
                first = data[0]
                try:
                    lat = float(first.get("lat"))
                    lon = float(first.get("lon"))
                    return lat, lon
                except Exception:
                    return None, None
            else:
                # no matches
                return None, None

        if resp.status_code in (429, 503):
            # rate limited or service unavailable - backoff and retry
            if attempt == max_retries:
                return None, None
            time.sleep(backoff)
            backoff *= 2
            continue

        # for 403/4xx/5xx other than 429/503 -> don't retry too aggressively
        # if 403, you might be blocked/need to contact API provider
        return None, None

    return None, None


def geocode_addresses(df: pd.DataFrame,
                      street_col: str,
                      city_col: Optional[str],
                      state_col: Optional[str],
                      postal_col: Optional[str],
                      api_key: Optional[str] = None,
                      country: str = "US",
                      timeout: float = 10.0,
                      pause: float = 1.0,
                      max_retries: int = 4,
                      max_rows: Optional[int] = None,
                      show_progress: bool = True) -> pd.DataFrame:
    """
    Geocode addresses from a DataFrame and return a copy with 'latitude_geocoded'
    and 'longitude_geocoded' columns appended.

    Example:
      df2 = geocode_addresses(df, "street_col", "city_col", "state_col", "zip_col", api_key="XXX")
      # then df2 has new columns you can assign back: df["lat"] = df2["latitude_geocoded"]
    """
    session = requests.Session()
    cache: Dict[str, Tuple[Optional[float], Optional[float]]] = {}

    n = len(df) if max_rows is None else min(len(df), max_rows)
    latitudes = []
    longitudes = []

    it = range(n)
    for i in it:
        if show_progress and (i % 50 == 0):
            print(f"Geocoding row {i+1}/{n} ...")
        row = df.iloc[i]
        street = row.get(street_col, None)
        city = row.get(city_col, None) if city_col else None
        state = row.get(state_col, None) if state_col else None
        postal = row.get(postal_col, None) if postal_col else None

        addr_key = _build_address_string(street, city, state, postal)
        if addr_key in cache:
            lat, lon = cache[addr_key]
        else:
            lat, lon = geocode_single(session,
                                      street=street,
                                      city=city,
                                      state=state,
                                      postalcode=postal,
                                      api_key=api_key,
                                      country=country,
                                      timeout=timeout,
                                      max_retries=max_retries,
                                      pause=pause)
            cache[addr_key] = (lat, lon)
            # polite pause between requests to avoid being rate-limited
            time.sleep(pause)

        latitudes.append(lat if lat is not None else np.nan)
        longitudes.append(lon if lon is not None else np.nan)

    # Build output DataFrame (copy to avoid mutating input)
    out = df.copy()
    # New columns are aligned with the subset length; if max_rows set, fill rest with NaN
    if max_rows is not None and max_rows < len(df):
        # create full-length lists
        lat_full = [np.nan] * len(df)
        lon_full = [np.nan] * len(df)
        for j in range(max_rows):
            lat_full[j] = latitudes[j]
            lon_full[j] = longitudes[j]
        out["latitude_geocoded"] = lat_full
        out["longitude_geocoded"] = lon_full
    else:
        out["latitude_geocoded"] = latitudes + [np.nan] * (len(df) - len(latitudes))
        out["longitude_geocoded"] = longitudes + [np.nan] * (len(df) - len(longitudes))

    return out
```
